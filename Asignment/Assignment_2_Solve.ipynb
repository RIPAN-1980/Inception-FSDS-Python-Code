# Problem - 1: Email Validator (String Filtering + Loops):
üíº Real-world use: Used in signup forms or data cleaning tasks.
Task: Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù

Example:
Input: hello@inceptionbd.com
Output: Valid Email

```python
email_id = input("enter your valid email id: ")

has_at = False
end_with = False

for ch in email_id:
  if ch == "@":
    has_at = True
    break

if email_id.endswith (".com") or email_id.endswith (".org"):
  end_with = True
if has_at and end_with:
  print("Valid Email")
else :
  print("Invalid Email Id.")
```

# Problem-2. Count Word Frequency in Customer Reviews:

üíº Real-world use: Analyzing reviews or comments in text analytics.
Task: Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
customers_review = input("please enter your review: ")
customers_review = customers_review.lower()
words = customers_review.split()
word_count = {}

for word in words:
  if word in word_count:
    word_count[word] += 1
  else:
    word_count[word] = 1

print("Word Frequency Count: ")
print(word_count )
```

# Problem-3. Password Strength Checker
üíº Real-world use: Used in user authentication systems.
Task: Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one special character.
Example:
Input: "Hello@123"
Output: Strong Password

```python
password = input("Enter Your Valid Password: ")
special_charecter = "!@#$%^&*()-_=+[]{};:,./<>?"
has_upper = False
has_lower = False
has_digit = False
has_special_chr = False

for ch in password:
  if ch.isupper():
    has_upper = True
  elif ch.islower():
    has_lower = True
  elif ch.isdigit():
    has_digit = True
  elif ch in special_charecter:
    has_special_chr = True

if has_upper and has_lower and has_digit and has_special_chr:
  print("Strong Password")
else:
  print("Week Passeord")
```

## Problem _4. Username Generator
üíº Real-world use: Automatically create usernames for users.
Task: Take user‚Äôs full name and generate a short username. Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).

Example:
Input: "Mamun Malitha"
Output: mamitha57

```python
name = input("enter your full name: ")
full_name = name.lower() .split()
import random

first_name = full_name[0]
last_name = full_name[-1]

first_3_ch = first_name
first_3 = ""
i = 0
while i < 3:
    first_3 += first_3_ch[i]
    i += 1

last_3_ch = last_name
last_3 = ""
length = len(last_3_ch)
i = length - 3
while i < length:
    last_3 += last_3_ch[i]
    i += 1

num = ""
for _ in range(2):
    num += str(random.randint(0, 9))

print(first_3 + last_3 + num)
```

Task: Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).

Input: "This is a demo message."

```python
sms = input("Type Your Massage: ")
sms_word = sms.split()
sms_ch_count = len(sms)
sms_word_count = len(sms_word)

if sms_ch_count>160:
  print("Exceed SMS Limit.")
else:
  print("Status = Within Limit.")

print("Characters:", sms_ch_count)
print("Words:", sms_word_count)
```

# Problem - 6: Remove Duplicates from Contact List
üíº Real-world use: Data cleaning in CRM or user databases.
Task: Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"

```python
contacts = input("Enter contacts separated by commas: ")  # Karim, Rahim, Jadeb, Rahim, Karim, Milon
contact_list = contacts.split(",")

unique_name = []

for i in contact_list:
    i = i.strip()
    if i not in unique_name:
        unique_name.append(i)

print("Unique Contacts:", unique_name)
```

# Problem - 7: Detect Suspicious Words in Text (Moderation)
üíº Real-world use: Used in chat moderation or email filtering.
Task: If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.
Example:
Input: "This is a scam offer"
Output: Warning: Suspicious content detected

```python
text = input("Enter Your Massege: ")
ban_word = ["spam", "scam", "fake"]
text_list = text.lower()
ban_word_found= False

for i in ban_word:
  if i in text_list:
    ban_word_found = True
    break

if ban_word_found:
  print("Warning: Suspicious content detected")
else:
  print("Massage is safe.")
```

# Problem_9: Count Hashtags and Mentions (Social Media Analytics)
üíº Real-world use: Used in analyzing social media posts.
Task: Count how many hashtags (#) and mentions (@) appear in a post.
Example:
Input: "Check out #AIWorkshop and follow @inceptionbd"
Output: Hashtags = 1, Mentions = 1

```python
post = input("Write your post here. :")
hashtag_count = 0
mention_count = 0

for ch in post:
  if ch == "#":
    hashtag_count += 1
  elif ch == "@":
    mention_count += 1


print("hashtag:", hashtag_count)
print("Menion: ", mention_count)
```
