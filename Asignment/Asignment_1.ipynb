## Problem - 1: Temperature Check
"""Write a program that takes the current temperature as input and prints:
● "It's too cold!" if below 10°C
● "It's cool outside" if between 10°C and 25°C
● "It's hot outside!" if above 25°C """

```python
temp = float((input("Enter the current tempeture in 0c:" ) ))
if temp < 10:
  print("It's too cold")
elif temp >= 10 and temp <=25:
  print("It's cool outside")
else:
  print("it's hot outside")
```

## Problem - 2: ATM Withdrawal System

"""
Problem: Ask the user for their account balance and the amount they want to withdraw.
● If the withdrawal amount is greater than the balance → print "Insufficient Balance"
● If the withdrawal amount is less than or equal to balance → print "Transaction Successful" and show the remaining balance.
"""

```python
balance = float((input("Enter your account balance: ")))
withdrawal_amount = int(input("Amount you want to withdraw: "))
remaining_balance = (balance- withdrawal_amount)
if withdrawal_amount > balance:
  print("Insufficient balance")
else :
  print("Transaction Successful, Remaining Balance: ", remaining_balance )
```

## Problem - 3: Student Grade Calculator

```python
marks = float(input("Enter your mark within 0 to 100 range: "))
if marks >=90 and marks <=100:
  print("Grade = A+")
elif marks >=80 and marks <=89:
  print("Grade = A")
elif marks >=70 and marks <=79:
  print("Grade = B")
elif marks >=60 and marks <=69:
  print("Grade = C")
else:
  print("Fail")
```

##Problem - 8: Leap Year Checker

 """Take a year as input and check:
● If it is divisible by 400 → Leap Year
● Else if divisible by 4 but not by 100 → Leap Year
● Otherwise → Not a Leap Year"""

```python
year = int(input("input the taken year: "))

if year % 4 == 0 and year % 100 != 0:
  print("Leap Year")
elif year % 400 == 0:
  print("Leap Year")
else:
  print("Not a Leap Year")
```
