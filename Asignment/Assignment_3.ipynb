# **Problem-1: Inventory Price Calculator**

**Real-world use:** Used in e-commerce dashboards for price summaries.

**Task:**
Write a function calculate_inventory_summary (prices) that:

 receives a list of product prices

 returns total_cost and average_cost

**Example:**

**Input**: [120, 250, 399, 150]

**Output**: (919, 229.75)

```python
#def calculate_inventory_summary (total_cost, avg_cost):
def calculate_inventory_summary (prices):
  prices = [120, 132, 230, 321]
  total_numbers = len(prices)
  total_cost = sum(prices)
  avg_cost = total_cost / total_numbers
  return (total_cost , avg_cost)

print("Total Cost: ", total_cost, "; " "Average Cost: ", avg_cost,)
```

```python
# Call Function
prices = [100, 200, 300, 200, 200]
calculate_inventory_summary (prices)
```

# **Problem-2: Unique Visitor Tracker**

**Real-world use:** Website analytics tools calculating daily active users.

**Task:**
Write a function count_unique_visitors(visitor_list) that:

 takes a list of visitor IDs

 returns the number of unique visitors

**Example:**

**Input:**  [101, 205, 101, 310, 205, 550]

**Output:**  4 (The unique IDs are 101, 205, 310, 550)

```python
def count_unique_visitors(visitor_list):
  visitor_list = [101, 205, 101, 310, 205, 550]
  unique_ids = set(visitor_list)
  number_of_unique_ids = len(unique_ids)

  #return('Number of unique ids:', number_of_unique_ids, "Unique ids are:", unique_ids)

  return(number_of_unique_ids, unique_ids)

print("Number of unique ids:", number_of_unique_ids)
print("Unique ids are:" , unique_ids)
```

```python
# Call Function
visitor_list = [101, 205, 101, 310, 205, 550, 205]
count_unique_visitors(visitor_list)
```

# **Problem-3: Product Category Counter**

**Real-world use:** Inventory/category-based analytics

**Task:**
 Write a function count_categories(products) that:

 takes a list of product names

 returns a dictionary counting each product type.

**Example:**

**Input:** ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]

**Output:** {"Laptop": 3, "Smartphone": 2, "Monitor": 1, "Headphones": 1}

```python
def count_categories(products):
  products = ["Laptop" , "Smartphone", "Monitor", "Laptop", "Smartphone","Headphones", "Laptop", "Smartphone", "Monitor", "Laptop"]
  products_count = {}
  for items in products:
    if items in products_count:
      products_count[items] += 1
    else:
      products_count[items] = 1
  return(products_count)

count_categories(products)
#print(products_count)
```

# **Prombel-4: Temperature Analytics**

**Real-world use: Health tech temperature screening systems.**

**Task:**
 Write a function analyze_temperatures(temp_list) that:

 separates high (> 38°C) and normal temps

 returns two lists: (high_list, normal_list)

**Example:**

**Input:** [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]

**Output:**

High List (>38°C): [39.1, 40.5]

Normal List (<38°C): [36.5, 38.0, 37.2, 38.0, 37.9]

Full Output Tuple: ([39.1, 40.5], [36.5, 38.0, 37.2, 38.0, 37.9])

```python
def analyze_temperatures(temp_list):
  temp_list = [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]
  high_temp_list = []
  normal_temp_list = []

  for tem in temp_list:
    if tem > 38:
      high_temp_list.append(tem)
    else:
      normal_temp_list.append(tem)
  return (high_temp_list, normal_temp_list)

print(high_temp_list) # why high_temp_list show {} blank item
print(normal_temp_list)

analyze_temperatures(temp_list)
```
